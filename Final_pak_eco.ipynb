## 3.Data Cleaning

```python
# Convert 'created_at' to datetime
pk['created_at'] = pd.to_datetime(pk['created_at'])

```

```python
# Check if 'created_at' and 'Working_Date' are the same for each order
pk['date_match'] = pk['created_at'] == pk['Working Date']
```

```python
# Count the number of orders where 'created_at' and 'Working_Date' are different
mismatched_count = (pk['date_match'] == False).sum()
```

```python
# Print the count of mismatched orders
print(f"There are {mismatched_count} orders with different 'created_at' and 'Working_Date'.")


```

```python
# Remove redundancy by deleting one of the columns
# In this example, we will keep 'created_at' and rename it to 'order_date'
pk.drop('Working Date', axis=1, inplace=True)
```

The code checks if the 'created_at' and 'Working_Date' columns in the dataset hold the same date information for each order. It identifies no differences, meaning both columns are consistent. After confirming this match, the code then removes redundancy by deleting the 'Working_Date' column and retains 'created_at' as the date reference for orders. This ensures data consistency and removes unnecessary duplicate information.


```python
pk.rename(columns={'created_at': 'order_date'}, inplace=True)
```

```python
print(pk['order_date'])
```

```python
pk['order_date'] = pd.to_datetime(pk['order_date'])
```

The code renames the 'created_at' column to 'order_date' and converts the values within the 'order_date' column to a datetime format. The printed output displays the 'order_date' column with its values transformed into a consistent date-time format for each order.

```python
pk.duplicated()
pk = pk.drop_duplicates(keep='first')
```

The code inspects the DataFrame for duplicated rows. Once identified, the duplicate rows are removed while keeping the first occurrence of the duplicated entries. This ensures that the dataset contains only unique rows, and the output shows the DataFrame with any identified duplicates removed.

```python
#conversion of dtypes
pk["item_id"] = pk["item_id"].astype(int)
pk["qty_ordered"] = pk["qty_ordered"].astype(int)
pk["grand_total"] = pk["grand_total"].astype(int)
pk["Year"] = pk["Year"].astype(int)
pk["Customer ID"] = pk["Customer ID"].astype(str)
pk["Month"] = pk["Month"].astype(int)

```

This code is modifying the data types of specific columns within the DataFrame 'pk' to enhance their usability or fix inconsistencies. It converts 'item_id', 'qty_ordered', and 'grand_total' columns to integers, 'Year' and 'Month' to integers, and 'Customer ID' to string format

This line of code fills in the missing values in the 'sales_commission_code' column in the DataFrame 'pk' with the string 'No Commission code'. By employing the fillna() function, it modifies the DataFrame in place ('inplace=True'), ensuring the changes are applied directly to the DataFrame 'pk'. This can be helpful in handling null values, replacing them with a designated value to maintain data integrity and analysis accuracy.

This line of code fills the missing values in the 'category_name_1' column using the 'forward fill' method ('method=ffill'), where the last valid observation is propagated forward to fill the gaps. This technique ensures missing values are replaced with the most recent non-null value in the column.

```python

# Assuming 'pk' is your DataFrame containing the relevant data with columns: 'category_name', 'qty_ordered', and 'price'

# Calculate total sales volume and revenue by category
category_sales = pk.groupby('category_name').agg({'qty_ordered': 'sum', 'price': 'sum'})
category_sales['revenue'] = category_sales['qty_ordered'] * category_sales['price']

# Sort the data by sales volume in descending order
category_sales = category_sales.sort_values('qty_ordered', ascending=False)

# Plotting the data
plt.figure(figsize=(12, 8))

# Bar plot for sales volume by category
plt.subplot(2, 1, 1)
category_sales['qty_ordered'].plot(kind='bar', color='blue')
plt.title('Sales Volume by Category')
plt.xlabel('Category')
plt.ylabel('Sales Volume')
plt.xticks(rotation=45)

# Bar plot for revenue by category
plt.subplot(2, 1, 2)
category_sales['revenue'].plot(kind='bar', color='blue')
plt.title('Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

The category that exhibits the highest sales volume and revenue generated is "Mobile and Tablets".

##percentage for top selling category by grant total

```python


# Group the data by category and calculate the total grand total for each category
G_total = pk.groupby(["category_name"], as_index=False)["grand_total"].sum().sort_values(by="grand_total", ascending=False)

plt.figure(figsize=(16, 6))
plt.xticks(rotation=45)

# Create the bar plot with blue bars
fig = sns.barplot(data=G_total, x="category_name", y="grand_total", color='blue')

plt.xlabel("Name of Category", size=15)
plt.ylabel("Grand Total", size=15)
plt.title("Name of Category vs Grand Total", size=15)

# Calculate the total grand total for all categories
total = sum(G_total["grand_total"])

# Add percentage labels on top of the bars
for i, v in enumerate(G_total["grand_total"]):
    percentage = v / total * 100
    plt.text(i, v, f'{percentage:.1f}%', ha='center', va='bottom')

plt.show()
```

This insight reveals that a substantial portion, 48.9%, of the total revenue comes from the sales of mobile and tablet categories.

```python

# Convert 'order_date' to datetime format if not already
pk['order_date'] = pd.to_datetime(pk['order_date'])

# Define the date range for Black Friday in 2017
black_friday_start = '2017-11-24'
black_friday_end = '2017-11-24'


# Filter data for Black Friday
black_friday_data = pk[(pk['order_date'] >= black_friday_start) & (pk['order_date'] <= black_friday_end)]

# Filter data for the "Mobiles & Tablets" category
mobiles_tablets_data = black_friday_data[black_friday_data['category_name'] == "Mobiles & Tablets"]

# Calculate total sales for Black Friday in the "Mobiles & Tablets" category
total_sales_black_friday = mobiles_tablets_data['grand_total'].sum()

# Filter data for the remaining days in November 2017
rest_of_november_data = pk[(pk['order_date'] >= '2017-11-01') & (pk['order_date'] <= '2017-11-30')]

# Filter data for the "Mobiles & Tablets" category
rest_of_november_mobiles_tablets_data = rest_of_november_data[rest_of_november_data['category_name'] == "Mobiles & Tablets"]

# Calculate the average sales for the "Mobiles & Tablets" category in November 2017
average_sales_november = rest_of_november_mobiles_tablets_data['grand_total'].mean()

# Compare the total sales on Black Friday with the average sales for the rest of November
if total_sales_black_friday > average_sales_november:
    print("Black Friday in 2017 had higher sales in the 'Mobiles & Tablets' category compared to the average sales for the rest of November.")
else:
    print("Black Friday in 2017 did not have higher sales in the 'Mobiles & Tablets' category compared to the average sales for the rest of November.")
```
